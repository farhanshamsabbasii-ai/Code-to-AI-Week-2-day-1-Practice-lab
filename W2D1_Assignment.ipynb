### The task
Tum apne kisi bhi pasandeeda topic ki **"Top 5" list** banao — e.g. top 5 movies, top 5 foods, top 5 cricketers, top 5 games. Phir us par kuch operations karo (neeche steps diye hain).


## Step 1 — Make your Top 5 list
Apni 5 cheezein ek list mein daalo aur poori list print karo.

````python
# TODO: apni top 5 list banao
top5 = ["Farhan", "loves", "Biryani", "too", "much"]   # e.g. ["Babar Azam", "Virat Kohli", ...]

print(top5)
````

## Step 2 — Show count
List mein kitne items hain, print karo using `len()`.

````python
# TODO: print how many items
print(len(top5))
````

## Step 3 — Print nicely with a loop ⭐
`for` loop use karke har item ko apni line par print karo, aage ek dash (`-`) ke saath.
Expected style:
```
- item one
- item two
...
```

````python
# TODO: loop over top5 and print "- <item>" for each
for item in top5:
    print("-" , item)
````

## Step 4 — First, last & sorted
1. Pehla item print karo (index 0)
2. Aakhri item print karo (index -1)
3. Poori list ko **sorted** form mein print karo (use `sorted()`)

````python
# TODO: first item
print(top5[0])

# TODO: last item
print(top5[-1])

# TODO: sorted list
print(sorted(top5))
````

## Step 5 — Edit your list
1. Ek naya item `append()` se add karo (ab top 6 ho gaye 😄)
2. Koi ek item `remove()` se nikaalo
3. Final list print karo

````python
# TODO: append one item
top5.append(136)

# TODO: remove one item
top5.remove("too")

# TODO: print final list
print(top5)
````

## Step 6 — A tuple fact 🔒
Apne baare mein ek **tuple** banao jo change na ho: `(your_name, your_city)`.
Phir loop chala kar dono values print karo.

````python
# TODO: make your info tuple
my_info = ("Farhan" , "Karachi")

# TODO: loop over it and print each value
for info in my_info:
    print(info)
````

## 🌟 Bonus (optional)
Do lists banao — `subjects` aur `marks` (same length).
`subjects` ko sort karke print karo, aur `marks` list ka use karke har mark ko loop se print karo.

*(Koi if/else nahi — sirf loops, sort, len.)*

````python
subjects = ["Math", "English", "Science", "Urdu"]
marks = [85, 78, 92, 88]

# TODO: sorted subjects
subjects.sort()
print(subjects)

# TODO: loop and print each mark
for mark in marks:
    print(mark)
````

## Exercise 2 — Slicing
Isi `numbers` list se:
1. Pehle 3 numbers
2. Aakhri 2 numbers
3. Beech ke (index 2 se 4 tak)

```python
numbers = [5, 10, 15, 20, 25, 30, 35]

# TODO: first 3
print(numbers[0:3])
print(numbers[-7:-4]) # or

# TODO: last 2
print(numbers[5:7])
print(numbers[-2:]) # or

# TODO: index 2 to 4
print(numbers[2:4])
print(numbers[-5:-3]) # or
```

## Exercise 3 — Changing a list

## Exercise 5 — ⭐ Your first loops
Yeh hai is week ka star! `for` loop use karo.

## Exercise 6 — Tuples
1. Ek tuple banao `(width, height)` = `(1920, 1080)`
2. Width aur height alag-alag print karo
3. Tuple ko loop karke dono values print karo

```python
# TODO: make the tuple
screen = (1920 , 1080)

# TODO: print width and height
print("Width:" , screen[0])
print("Height:" , screen[1])

# TODO: loop over the tuple
for value in screen:
    print(value)
```

## 🌟 Stretch challenge (optional, thoda mushkil)
Ek list `temps = [31, 28, 35, 40, 26]` di hai.
- Loop chala kar har temperature ko `"<temp>°C"` format mein print karo.
- Phir list ko sort karke print karo (sabse thandi se garam tak).

*(Sirf loop aur sort — koi if/else nahi!)*

```python
temps = [31, 28, 35, 40, 26]

# TODO: loop and print each as "<temp>°C"
for temp in temps:
    print(temp, "⁰C")

# TODO: sorted version
temps.sort()

print(temps)
```
